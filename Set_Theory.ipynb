```python
from matplotlib_venn import venn3, venn3_circles
import matplotlib.pyplot as plt

# Given data
newsweek_only = 65 - (20 + 25 - 8)
times_only = 45 - (20 + 15 - 8)
fortune_only = 42 - (25 + 15 - 8)
newsweek_and_times = 20 - 8
newsweek_and_fortune = 25 - 8
times_and_fortune = 15 - 8
all_three = 8

#Let X be the people who read at least one of the three magazines
X = 65 + 45 + 42 - 25 - 20 - 15 + 8

#Let Y be the total of people who have read exactly one magazine
fortune_readers = 42 - 25 - 15 + 8
newsweek_readers = 65 - 25 - 20 + 8
times_readers = 45 - 15 -20 + 8
Y = fortune_readers + newsweek_readers + times_readers

# Create the Venn diagram
venn = venn3(subsets=(newsweek_only, times_only, newsweek_and_times,
                      fortune_only, newsweek_and_fortune, times_and_fortune, all_three),
             set_labels=('Newsweek', 'Times', 'Fortune'))

# Draw circles for the universal set
venn_circles = venn3_circles(subsets=(newsweek_only, times_only, newsweek_and_times,
                                      fortune_only, newsweek_and_fortune, times_and_fortune, all_three))
venn_circles[0].set_lw(1.0)  # Set line width for circle 1
venn_circles[1].set_lw(1.0)  # Set line width for circle 2
venn_circles[2].set_lw(1.0)  # Set line width for circle 3

plt.title("Venn Diagram of Magazines")
plt.show()

print("Number of people who read at least one of the 3 magazines:", X)
print("Total number of people who read at exactly one magazine:", Y)
```

```python
from matplotlib_venn import venn3, venn3_circles
import matplotlib.pyplot as plt

# Given data
T = 25
AW = 5
AR = 9
RW = 4
ARW = 3

# Calculating the values of A, R, and W based on the given data
A = 15 - (AW + AR - ARW)
R = 12 - (RW + AR - ARW)
W = 11 - (RW + AW - ARW)

AW = 5 - ARW
AR = 9 - ARW
RW = 4 - ARW

#Let O be the number of cars with at least one option
O = 15 + 11 + 12 - 5 - 9 - 4 + 3

# Let N be the number of cars with none of the options
N = T - (15 + 11 + 12 - 5 - 9 - 4 + 3)

# Create the Venn diagram
venn = venn3(subsets=(A, R, AR, W, AW, RW, ARW), set_labels=('A', 'R', 'W'))

# Create a larger universal set circle labeled "U" with the value 1
plt.annotate("N\n2", xy=(0.5, 0.5), fontsize=10, ha='center', va='center', color='black', weight='bold')

# Draw circles for the universal set
venn_circles = venn3_circles(subsets=(A, R, AR, W, AW, RW, ARW))
venn_circles[0].set_lw(1.0)  # Set line width for circle 1
venn_circles[1].set_lw(1.0)  # Set line width for circle 2
venn_circles[2].set_lw(1.0)  # Set line width for circle 3

plt.title("Venn Diagram of Car Features")
plt.show()

# Print the answers
print("Number of cars with W only:", W)
print("Number of cars with A only:", A)
print("Number of cars with R only:", R)
print("Number of cars with both R and W but not A:", RW)
print("Number of cars with both A and R but not W:", AR)
print("Number of cars with only one of the options:", A+R+W)
print("Number of cars with at least one option:", O)
print("Number of cars with none of the options:", N)
```
